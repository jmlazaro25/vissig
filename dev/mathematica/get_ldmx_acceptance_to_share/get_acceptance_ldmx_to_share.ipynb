# Acceptance for Visible Decays inside of LDMX (8 GeV or 4 GeV, 1.2 GeV Cut)

```python

# The default LDMX configuration at 4 GeV
zmin = 28. #7.
zmax = 300.
det_rad = 100.
Ebeam = 4.
Ethr = 1.2

fout = './LDMX_visible_acceptance_grid_ma_vs_ctau_Ebeam_4GeV_zmin_28cm_zmax_300cm_Ethr_1pt2GeV.dat'

xsec_and_pz = np.loadtxt('./xsec_and_pz_list_4gev_recoil_cut_1p2GeV.dat')
pz_interp = interpolate.interp1d(xsec_and_pz[:,0],xsec_and_pz[:,2])

dir_path = '/home/nblinov/Documents/Projects/alptraum/ldmx_ap_minimal_Ebeam_4_GeV_no_decay/Events/'

```

```python
start_time = time.time()

ctau_full_range = np.logspace(-4., 6., 100)



run_list = [d for d in os.listdir(dir_path) if os.path.isdir(os.path.join(dir_path, d))]
#run_list = ['ma_162']

gct_max = 100.*zmax
gct_min = zmin/100.

acceptance_grid = []

for run in run_list:
    rd = dir_path + run
    print "Working on " + rd
    
    # Determine a reasonable epsilon range
    ma = get_mass_from_pc(rd,pid=622, banner_tag=banner_tag)
    boost_avg = pz_interp(ma)/ma
    
    # Determine a reasonable ctau range
    ctau_min = gct_min/boost_avg
    ctau_max = gct_max/boost_avg
    ctau_list = np.logspace(np.log10(ctau_min),np.log10(ctau_max),20)

    evt_weight_list, four_vector_list = get_momentum_list(rd, banner_tag=banner_tag)
    
    acceptance_subgrid = []
    for ctau in ctau_list:
        accept = get_acceptance_from_four_vectors(ma, ctau, four_vector_list, zmin, zmax, det_rad,Ethr=Ethr)
                  
        acceptance_subgrid.append([ma,ctau,accept])
        print "ma, ctau, acceptance = ", acceptance_subgrid[-1]
    
    acceptance_subgrid = sorted(acceptance_subgrid, key=lambda x: (x[0],x[1]))
    acceptance_subgrid = np.array(acceptance_subgrid)
    
    log_interpolation = interpolate.interp1d(np.log10(acceptance_subgrid[:,1]), np.log10(acceptance_subgrid[:,2]))

    def intermediate_ctau_interpolation(ctau):
        return 10**log_interpolation(np.log10(ctau))

    def large_ctau_extrapolation(ctau):
        return acceptance_subgrid[-1,2]*(acceptance_subgrid[-1,1]/ctau)

    def fit_func(ctau, a, b):
        return a - b*(zmin/ctau)

    fit_result = optimize.curve_fit(fit_func, acceptance_subgrid[::4,1],\
            np.log(acceptance_subgrid[::4,2]))

    def small_ctau_extrapolation(ctau):
        return np.exp(fit_func(ctau, fit_result[0][0], fit_result[0][1]))

    def full_range_interpolation(ctau):
        if ctau > min(acceptance_subgrid[:,1]) and ctau < max(acceptance_subgrid[:,1]):
            return intermediate_ctau_interpolation(ctau)
        elif ctau < min(acceptance_subgrid[:,1]):
            return small_ctau_extrapolation(ctau)
        else:
            return large_ctau_extrapolation(ctau)

    interpolated_grid = [[ma,ctau,full_range_interpolation(ctau)] for ctau in ctau_full_range]
    acceptance_grid.extend(interpolated_grid)
    interpolated_grid = np.array(interpolated_grid)
    
    plt.xscale('log')
    plt.yscale('log')
    plt.ylim([1e-10,4])
    plt.plot(acceptance_subgrid[:,1],acceptance_subgrid[:,2],'o')
    plt.plot(interpolated_grid[:,1],interpolated_grid[:,2])
    plt.title(run)
    plt.show()
    

acceptance_grid_sorted = sorted(acceptance_grid, key=lambda x: (x[0],x[1]))
acceptance_grid_sorted = np.array(acceptance_grid_sorted)

print "Writing output to ", fout
np.savetxt(fout, acceptance_grid_sorted)
print "Elapsed time: ", (time.time()-start_time)/60., " minutes."
```
